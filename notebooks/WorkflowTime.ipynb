```python
import pandas as pd
import numpy as np
```

```python
arc_log_fn = "/Users/nathanneeteson/Documents/Data/TRIKNEE/arc_logs.xlsx"
df = pd.concat(pd.read_excel(arc_log_fn, sheet_name=None, skiprows=13, skipfooter=1), ignore_index=True)

df.dropna(inplace=True)

df
```

```python
print(set(df["Job Name"]))
```

```python
def convert_to_seconds(x: str) -> int:
    if x[-1] == "s":
        return int(x[:-1])
    elif x[-1] == "m":
        return 60 * int(x[:-1])
    elif x[-1] == "h":
        return 60 * 60 * int(x[:-1])
    else:
        return np.nan


df["time"] = df["Runtime"].apply(
    lambda x:
    convert_to_seconds(x.split(" ")[0].strip())
)

df
```

```python
for job_name in set(df["Job Name"]):
    print(job_name)
    print(df[df["Job Name"]==job_name]["time"].mean() / 60)
    print(df[df["Job Name"]==job_name]["time"].std() / 60)
    print()
```
